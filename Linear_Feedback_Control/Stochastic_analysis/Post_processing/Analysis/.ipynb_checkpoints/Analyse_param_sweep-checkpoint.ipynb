```python
def compute_vh_gradients(x, max_t_idx):
    data = x.reset_index()
    dt = data.iloc[1]['t']-data.iloc[0]['t']
    
    grad_vh_sim = np.mean(np.gradient(data['vh_sim'].iloc[0:max_t_idx],dt))
    grad_vh_sse = np.mean(np.gradient(data['vh_sse'].iloc[0:max_t_idx], dt))
    grad_vh_ansatz = np.mean(np.gradient(data['vh_ansatz'].iloc[0:max_t_idx], dt))
    
    error_sim_sse = np.abs(1. - grad_vh_sse/grad_vh_sim)
    error_sim_ansatz = np.abs(1. - grad_vh_ansatz/grad_vh_sim)
    
    return pd.Series({'grad_vh_sim':grad_vh_sim,
                        'grad_vh_sse':grad_vh_sse,
                        'grad_vh_ansatz':grad_vh_ansatz,
                        'error_sim_sse':error_sim_sse,
                        'error_sim_ansatz':error_sim_ansatz})
    
    
```

```python
errors = vh_df.groupby('block_idx').apply(lambda x: compute_vh_gradients(x,10))
errors.head()
```

## Look at copy number CV as a function of $b$

```python
def cv_mom_clos(b,mu,wss):
    return np.sqrt(8*mu/b+2*wss*(wss-np.sqrt(wss*wss-8*mu/b)))/(3*wss+np.sqrt(wss*wss-8*mu/b))
```

```python
cv_b = []
for i in range(121,len(perturbed_params)):
    try:
        stats_data = pd.read_csv('../Data/online_stats_{}.csv'.format(i))
    except IOError:
        print('No stats found')
        continue # go to next iteration
    
       
    
    varn_fin = stats_data['var_n'].iloc[-1]
    meann_fin = stats_data['mean_n'].iloc[-1]
    
    cv_fin = np.sqrt(varn_fin)/meann_fin
    
    b_val = perturbed_params.iloc[i].b
    b_val_normed = np.log10(b_val/nominal_param[-3])
    
    cv_theory = cv_mom_clos(b_val, nominal_param[-2], meann_fin)
    
    cv_b.append([b_val_normed, cv_fin, cv_theory])
cv_b = np.array(cv_b)
    
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(cv_b[:,0],cv_b[:,1],'kx' , label = 'Simulation')
#ax.plot(cv_b[:,0],cv_b[:,2], '-r',label = 'Approx Theory')
ax.set_xlabel('Feedback control strength, $b$')
ax.set_ylabel('Copy number CV')
ax.legend()
plt.tight_layout()
plt.savefig('b_sweep.pdf')
```

The death rate is actually effectively lower than what I plugged into the eqn. Not sure the effect on the CV though. Anyway, it's the right order of magnitude, but is biased to overestimate the CV. Could get the SSE as better theory.
