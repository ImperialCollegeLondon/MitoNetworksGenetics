```python
c_file_param_ordering_convention = [
'xi',
'beta',
'gamma',
'kappa',
'b',
'mu',
'delta'
]
```

```python
hpc_workdir = '/work/ja1109/networks/sing_birth_fus/AJHG_ctrls/Y_linear_feedback_deg'
```

```python
c_filename = 'Y_linear_feedback_deg'
```

```python
hpc_jobname = 'Y_lfd'
```

```python
lfd = mts.SubmitHPCSoluableFeedbackControl(ss_definition=mtd.Y_linear_feedback_deg_ss,
                                 nominal_params=nominal_params,hpc_workdir=hpc_workdir,
                                c_filename=c_filename, c_file_param_ordering_convention=c_file_param_ordering_convention,
                                hpc_jobname=hpc_jobname,
                                         time_per_job = "23:00:00")
```
